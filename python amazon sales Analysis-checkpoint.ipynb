Steps 2-:Exploratory Data analysis

```python
df.columns
```

```python
ax=sns.countplot(x='Size',data=df)
#checking which size sold maximum
```

```python
plt.show()
```

```python
#for labels in bar graph
ax=sns.countplot(x='Size',data=df)
for i in ax.containers:
    ax.bar_label(i)
    
```

NOTE: From the above graph we can clearly see that gthe most of the people buys M size.

```python
df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
S_Qty=df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
ap=sns.barplot(x='Size',y='Quantity',data=S_Qty)
```

```python
ap=sns.barplot(x='Size',y='Quantity',data=S_Qty)
for p in ap.containers:
    ap.bar_label(p)
```

NOTE: From the above graph we can clearly see that the most of the Qty buys M size T-shirt.

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Courier Status',hue='Status',data=df)
plt.show()
```

FROM THE ABOVE GRAPH THE MAJORITY OF THE ORDERS ARE SHIPPED THROUGH COURIER.

```python
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=10,edgecolor='Black')
plt.show()
```

FROM the above graph we can clearly sees that most of the buyers are buying T-shirt.

```python
#checking the B2B by using Pie chart
B2B_CHECK=df['B2B'].value_counts()
plt.pie(B2B_CHECK,labels=B2B_CHECK.index,autopct='%1.1f%%')
plt.show()
```

FROM the above chart we can clearly see that the maximum i.e 99.2 % are Retailers and 0.8 % are B2B buyers.

```python
#  Prepare data for pie chart
a1 = df['Fulfilment'].value_counts()

# Step 4: Plot the pie chart
fig, ax = plt.subplots()

ax.pie(a1, labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect="equal")

plt.show()
```

FROM The above chart we can clearly see that most of the buyers are from "AMAZON",with 69.6 %.

```python
# Prepare data for scatter plot
x_data = df['Category']  
y_data = df['Size'] 

# Plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Category ')  
plt.ylabel('Size')  
plt.title('Scatter Plot') 
plt.show()
```

```python
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# top_10_States 
top_10_state = df['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```

Note: From above Graph you can see that most of the buyers are Maharashtra states.
